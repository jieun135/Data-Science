```python

#주피터 노트북안에 차트가 그려지도록 하기
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
```

```python

plt.plot(['Math','Korean','English'], [90, 100, 50])
plt.xlabel('Subjects')
plt.ylabel('Score')
plt.title('Midterm')

plt.plot(['Math','Korean','English'], [70, 60, 90])
plt.legend(['CHOI', 'KIM'])

plt.show()
```

```python
x = ['Math', 'Korean', 'English']
y = [90, 100, 50]

plt.xlabel('Subjects')
plt.ylabel('Score')
# plt.title('Midterm')
# plt.legend(['CHOI'])

plt.bar(x, y, width=0.7, color='pink')
plt.show()
```

```python
plt.plot([1,2,3], [90, 100, 50], marker='', markersize=10, linestyle=':')
# marker: o, v, ^, s, +, ., 빈문자
# linestyle: -, --, -., :, 빈문자

plt.grid(True) # 격자 그리기
plt.axis([0, 5, 50, 150]) # [x축 최소, x축 최대, y축 최소, y축 최대]

# plt.tight_layout()
plt.show()
```

```python
plt.plot([1,2,3],[90,100,50], marker = 's', markersize = 30, linestyle = '--')
plt.grid(True)
plt.axis([1,3,80,100])

plt.show()
```

판다스로 csv 데이터 불러오기

```python
import pandas as pd
```

```python
train = pd.read_csv('data/train.csv')

train
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
sns.set()
```

```python
sns.relplot(data = train, x = 'Age', y = 'Survived', hue = 'Sex', aspect =2)
```

```python
sns.catplot(data = train, x = 'Pclass', y = 'Fare', aspect = 4, jitter = False, hue = 'Sex')
```

```python
sns.pairplot(data = train, hue = 'Sex',
            x_vars=['Pclass', 'Age', 'Fare'],
            y_vars=['Survived','Age','Fare'],height=3)
```
