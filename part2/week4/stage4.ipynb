```python
import pandas as pd 

df = pd.read_csv('data/my_customers.csv')
df
```

```python
# 성별
df.loc[df['성별'] == 'male', '성별'] = 0
df.loc[df['성별'] == 'female', '성별'] = 1

# 관심사 
df.loc[df['관심사'] == '축구', '관심사'] = 0
df.loc[df['관심사'] == 'esports', '관심사'] = 1
df.loc[df['관심사'] == '야구', '관심사'] = 2
# 지역
df.loc[df['지역'] == '서울', '지역'] = 0
df.loc[df['지역'] == '경기', '지역'] = 1
df.loc[df['지역'] == '제주', '지역'] = 2

df
```

```python
input_data = df[['나이', '성별', '지역', '관심사']]
target_data = df['우수고객분류']
#train/valid 분ㄹ
## 다양한 모델을 실험해서 가장 좋은 모델 
```

```python
from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(n_estimators = 10)
forest.fit(input_data, target_data)

print('training set accuracy:', forest.score(input_data, target_data))
```

```python
df_new = pd.read_csv('data/new_customers.csv')
df_new
```

```python
prediction = forest.predict(x_test)
prediction
```
